```python

   #Given graph 
    #Giving input using dictionary
    
graph = {
  'Arad': {'Zerind': 75, 'Timisoara': 118, 'Sibiu': 140},
  'Bucharest': {'Urziceni': 85, 'Pitesti': 101, 'Giurgiu': 90, 'Fagaras': 211},
  'Craiova': {'Dobreta': 120, 'Rimnicu Vilcea': 146, 'Pitesti': 138},
  'Dobreta': {'Mehadia': 75, 'Craiova': 120},
  'Eforie': {'Hirsova': 86},
  'Fagaras': {'Sibiu': 99, 'Bucharest': 211},
  'Giurgiu': {'Bucharest': 90},
  'Hirsova': {'Urziceni': 98, 'Eforie': 86},
  'Iasi': {'Neamt': 87, 'Vaslui': 92},
  'Lugoj': {'Timisoara': 111, 'Mehadia': 70},
  'Mehadia': {'Lugoj': 70, 'Dobreta': 75},
  'Neamt': {'Iasi': 87},
  'Oradea': {'Zerind': 71, 'Sibiu': 151},
  'Pitesti': {'Rimnicu Vilcea': 97, 'Bucharest': 101, 'Craiova': 138},
  'Rimnicu Vilcea': {'Sibiu': 80, 'Craiova': 146, 'Pitesti': 97},
  'Sibiu': {'Arad': 140, 'Oradea': 151, 'Fagaras': 99, 'Rimnicu Vilcea': 80},
  'Timisoara': {'Arad': 118, 'Lugoj': 111},
  'Urziceni': {'Bucharest': 85, 'Hirsova': 98, 'Vaslui': 142},
  'Vaslui': {'Iasi': 92, 'Urziceni': 142},
  'Zerind': {'Arad': 75, 'Oradea': 71}
}

print("               *-*-*-GRAPH-*-*-*\n")
for key, value in graph.items():  #Printing Graph
    print(key, ' : ', value)

def BFS(graph, start, target):    #BFS function
    queue = [(start, [start])]
    while queue:                      #Uses Queue data structure
        (vertex, path) = queue.pop(0)
        for neighbor in graph[vertex]:
            if neighbor not in path:
                if neighbor == target:
                    return path + [neighbor]   
                else:
                    queue.append((neighbor, path + [neighbor]))   #Built-in func

def DFS(graph, start, target):    #DFS function
    stack = [(start, [start])]
    while stack:               #Uses Stack data structure
        (vertex, path) = stack.pop()
        for neighbor in graph[vertex]:
            if neighbor not in path:
                if neighbor == target:
                    return path + [neighbor]
                else:
                    stack.append((neighbor, path + [neighbor]))   #Built-in func

def comparePaths(DFS_path, BFS_path, graph):  #Path calculating func
    if DFS_path and BFS_path:
        dfs_dist = sum(graph[DFS_path[i]][DFS_path[i + 1]] for i in range(len(DFS_path) - 1))
        bfs_dist = sum(graph[BFS_path[i]][BFS_path[i + 1]] for i in range(len(BFS_path) - 1))
  
        if bfs_dist < dfs_dist:
          print(f"\nBFS outperforms DFS with total distance: {bfs_dist} Km")
        elif bfs_dist > dfs_dist:
          print(f"\nDFS outperforms BFS with total distance: {dfs_dist} Km")
        else:
          print(f"\nBoth have the same distance of: {bfs_dist} Km")
    else:
        print("No path exists between the given nodes in the graph.")                    

        
                    
start = 'Arad'   #Starting point
target = 'Bucharest'   #Ending point / Destination

#Gives the shortest path using BFS
bfs_path = BFS(graph, start, target)    #Calling func
print(f"\nThe shortest path from '{start}' to '{target}' using BFS is: {bfs_path}")

#Gives the shortest path using DFS 
dfs_path = DFS(graph, start, target)     #Calling func
print(f"\nThe shortest path from '{start}' to '{target}' using DFS is: {dfs_path}")

comparePaths(dfs_path, bfs_path, graph)   #Calling function
    #Prints results

```
